```python
import numpy as np
import matplotlib.pyplot as plt

plt.ion()
plt.title('Representacion de (x, f(x))')
plt.xlabel('Valores x')
plt.ylabel('Valores f(x)')
plt.ylim(0,1.3)
x = np.arange(50)
y = np.random.rand(50)
plt.plot(x,y,color='blue',label = 'x, f(x)')
plt.plot(x[y>0.7], y[y>0.7],'bo',label='f(x) > 0.7')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

lenguajes = ('Alemania', 'Espaa', 'Francia', 'Grecia', 'Paises bajos')
posicion_y = np.arange(len(lenguajes))
unidades = (72, 52, 21, 34, 83)
plt.barh(posicion_y, unidades, align = 'center')
plt.yticks(posicion_y, lenguajes)
plt.xlabel(' Porcentaje de usuarios de internet')
plt.title('Usuarios de internet ao 2007 por paises')
plt.grid(True)
plt.savefig('Usuarios de internet ao 2007 por paises.png')
plt.savefig('Usuarios de internet ao 2007 por paises.pdf')
```
